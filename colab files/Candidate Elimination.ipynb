```python
import numpy as np
import pandas as pd
data=pd.read_csv('/content/drive/MyDrive/finds.csv',names=[1,2,3,4,5,6,7])
concepts=np.array(data.iloc[:,:-1])
target=np.array(data.iloc[:,-1])
```

```python
def learn(concepts,target):
  specific_h=concepts[0].copy()
  general_h=[["?" for i in range(len(specific_h))]for i in range(len(specific_h))]
  for i,h in enumerate(concepts):
    if target[i]=="Yes":
      for x in range(len(specific_h)):
        if h[x]!=specific_h[x]:
          specific_h[x]="?"
          general_h[x][x]="?"
    if target[i]=="No":
      for x in range(len(specific_h)):
        if h[x]!=specific_h[x]:
          general_h[x][x]=specific_h[x]
        else:
          general_h[x][x]="?"
  indices=[i for i,val in enumerate(general_h) if val==['?','?','?','?','?','?']]
  for i in indices:
    general_h.remove(['?','?','?','?','?','?'])
  return general_h,specific_h

final_g,final_s=learn(concepts,target)
print("Final General Hypothesis is ",final_g)
print("Final Specific Hypothesis is ",final_s)
```
