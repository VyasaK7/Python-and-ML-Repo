# Wine dataset Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
data=pd.read_csv("/content/drive/MyDrive/DeepLearning_Workshop/winequality-red.csv",header=None)
print(data)
```

```python
wine_dataset=np.asarray(data)
print(wine_dataset)
print('\n')
naa=np.array(data)
print(naa)
```

```python
#setting input values
wine_dataset_features=wine_dataset[:,0:11]
print("Wine dataset features values are: \n",wine_dataset_features,"\n")
print("Shape of Wine dataset features is: \n",np.shape(wine_dataset_features),"\n")
```

```python
#seting target values
wine_dataset_target=wine_dataset[:,11]
print("Wine dataset target values are: \n", wine_dataset_target,"\n")
print("Shape of datase target values is: \n", np.shape(wine_dataset_target),"\n")
```

```python
#random state=42 always gives fixed order
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(wine_dataset_features,wine_dataset_target,test_size=0.05,random_state=42)
```

```python
print("Training data is \n",x_train, '\n')
print("Training data shape is \n",np.shape(x_train), '\n')
print("Training labels are \n",y_train, '\n')
print("Training labels shape is \n",np.shape(y_train), '\n')
```

```python
print("Testing data is \n",x_test, '\n')
print("Testing data shape is \n",np.shape(x_test), '\n')
print("Testing labels are \n",y_test, '\n')
print("Testing labels shape is \n",np.shape(y_test), '\n')
```

```python
from sklearn.linear_model import LinearRegression
linear_regression_model=LinearRegression(fit_intercept=True,normalize=True)
```

```python
print(linear_regression_model)
linear_regression_model.fit(x_train,y_train)
```

```python
print("Linear regression coefficients=",linear_regression_model.coef_,"\n")
print("Linear regression intercept=",linear_regression_model.intercept_)
```

```python
linear_regression_predictons=linear_regression_model.predict(x_test)
y_pred=linear_regression_predictons
print("The predictions of the linear regressor are: \n", linear_regression_predictons)
```

```python
from sklearn.metrics import r2_score
print("Accuracy score is: ",r2_score(y_test,y_pred))
#accuracy score gives error in linear regression(at least here it did)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test,linear_regression_predictons)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,linear_regression_predictons)
```

```python
print("Comparing the predictions with the Ground Truth \n",np.column_stack((y_test[:10],linear_regression_predictons[:10])))
r=r2_score(y_test,linear_regression_predictons)
print("r2 score is "+str(r))
```

# Allahabad Dataset

```python
dataset=pd.read_csv("/content/drive/MyDrive/DeepLearning_Workshop/allahabad.csv")
print(dataset)
```

```python
data_array=np.asarray(dataset)
print(data_array)
```

```python
dataset_features=data_array[:,0:10]
print("Dataset features values are: \n",dataset_features,"\n")
print("Shape of Dataset features is: \n",np.shape(dataset_features),"\n")
```

```python
#alternate way to create array of data
X=dataset['Day'].values
y=dataset['Confirmed'].values
```

```python
print(X.shape)
print(y.shape)
```

```python
print(X.shape)
print(y.shape)
```

```python
#print(X)
#print(y)
```

```python
regressor=LinearRegression()
#training model- fit used for training
regressor.fit(X,y)
```

```python
y_pred=regressor.predict(X)
```

```python
#using matpotlib
plt.scatter(X,y,color='Magenta')
plt.plot(X,y_pred,color='green')
plt.title("Daywise confirmed case in Allahabad")
plt.xlabel("No.Of.Day")
plt.ylabel("Confirmed cases")
plt.show()
```

# Social Network Logistic Regression

```python
df=pd.read_excel('/content/drive/MyDrive/DeepLearning_Workshop/Social_Network_Ads.xlsx')
print(df)
```

```python
df_getdummy=pd.get_dummies(data=df, columns=['Gender'])
df_getdummy
```

```python
df_getdummy1=df_getdummy.drop('User ID',axis=1)
```

```python
X=df_getdummy1.drop('Purchased',axis=1)
y=df_getdummy1['Purchased']
```

```python
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.10,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=0)
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score 
print("Confusion matrix for testing data is : \n",confusion_matrix(y_test,y_pred))
```

```python
print("Accuracy score for testing data is : \n",accuracy_score(y_test,y_pred))
```
