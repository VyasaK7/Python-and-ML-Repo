```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/content/drive/MyDrive/DeepLearning_Workshop/GBA.csv")
filtered_data = df.loc[df['SatisfactionCompany'].isin([1, 2, 3, 4])]
grouped_data = filtered_data.groupby(['SatisfactionCompany', 'LeftCompany'])['EmpID'].count().unstack()
#print(filtered_data.head())
print("XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX")
print(grouped_data.head())
```

```python
fig, ax = plt.subplots()
grouped_data.plot(kind='bar', stacked=True, ax=ax, color=['blue', 'orange'])
plt.title('Employee Job Satisfaction and Turnover')
plt.xlabel('Job Satisfaction Rating')
plt.ylabel('Number of Employees')
plt.xticks(range(4), ['1', '2', '3', '4'])
ax.legend(['Stayed', 'Left'], loc='upper left')
for i, bar in enumerate(ax.containers):
    for rect in bar:
        height = rect.get_height()
        ax.annotate(f'{int(height)}', xy=((rect.get_x() + rect.get_width() / 2),height), xytext=(0, 3),
                    textcoords='offset points', ha='center', va='bottom', fontsize=10)
plt.show()
```
