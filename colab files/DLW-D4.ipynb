# Naive Bayes Classifier

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
#load dataset
dataset=pd.read_csv("/content/drive/MyDrive/DeepLearning_Workshop/iris.csv")
print(dataset)
X = dataset.iloc[:,0:4].values
y = dataset.iloc[:,-1].values
```

```python
data = datasets.load_iris()
#print(data)
```

```python
np.shape(np.mat(data))
```

```python
# train and test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=10)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# Training Naive Bayes Model on Training Test
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(x_train,y_train)
```

```python
#Predicting the test results
y_pred=classifier.predict(x_test)
```

```python
#Determining Accuracy
from sklearn.metrics import confusion_matrix,accuracy_score
print("Confusion matrix for training data is : \n",confusion_matrix(y_train,classifier.predict(x_train))) 
print("Confusion matrix for testing data is : \n",confusion_matrix(y_test,y_pred))
print("Accuracy score for training data is : \n",accuracy_score(y_train,classifier.predict(x_train)))
print("Accuracy score for testing data is : \n",accuracy_score(y_test,y_pred))
```

# K Nearest Neighbors

```python
import numpy as np
from sklearn.datasets import load_wine 
```

```python
#Load and analyze the dataset
wine_dataset=load_wine()
#print(wine_dataset)
wine_dataset_features=wine_dataset.data
wine_dataset_labels=wine_dataset.target
print("Wine dataset feature values are: \n", wine_dataset_features , "\n")
print("Wine dataset labels are: \n", wine_dataset_labels , "\n")
```

```python
# train and test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(wine_dataset_features,wine_dataset_labels,test_size=0.25,random_state=28)
```

```python
#Analyse the training and testing sets
print("Training features is: \n", x_train, "\n")
print("Training labels are: \n", y_train, "\n")
print("Testing features is: \n", x_test, "\n")
print("Testing labels are: \n", y_test, "\n")
#Printing Shape
print("Shape of Training data is: \n", np.shape(x_train), "\n")
print("Shape of Training labels is: \n", np.shape(y_train), "\n")
print("Shape of Testing data is: \n", np.shape(x_test), "\n")
print("Shape of Testing labels is: \n", np.shape(y_test), "\n")
```

```python
#Loading the K-Nearest Classifier
from sklearn.neighbors import KNeighborsClassifier
knn_classifier=KNeighborsClassifier(n_neighbors=5)
print(knn_classifier)
```

```python
#Fitting the classifier on the training dataset
knn_classifier.fit(x_train,y_train)
```

```python
#Making predictions and comparision
knn_classifier_predictions=knn_classifier.predict(x_test)
print("KNN classifier predictions on the test data are: \n", knn_classifier_predictions,"\n")
print(np.shape(knn_classifier_predictions),"\n")
#ground truth comparision and getting score value
print("Comparing the Ground truth with the predictions [ground-truth knn-predictions]\n",np.column_stack((y_test,knn_classifier_predictions)),"\n")
print("KNN classifier score on the test dataset is: ",knn_classifier.score(x_test,y_test),"\n")
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print("Accuracy score for training data is : \n",accuracy_score(y_test,knn_classifier.predict(x_test)))
```

# Support Vector Machine

```python
#import libraries and dataset
import pandas as pd
import numpy as np
np.set_printoptions(suppress=True,precision=3)
path="/content/drive/MyDrive/DeepLearning_Workshop/iris.csv"
iris_dataset_frame = pd.read_csv(path,header=None)
print("Iris dataset dataframe looks like \n", iris_dataset_frame)
```

```python
#transforming the dataframe into numpy matrices
iris_dataset=iris_dataset_frame.to_numpy()
print("Iris Dataset: \n", iris_dataset, "\n")
print("np.shape(iris_dataset): ", iris_dataset.shape, "\n")
```

```python
#separating the features from labels in dataset
iris_dataset_features=iris_dataset[1:,0:4]
iris_dataset_labels=iris_dataset[1:,4]
print("np.shape(iris_dataset_features): ", np.shape(iris_dataset_features), "\n")
print("np.shape(iris_dataset_labels): ", np.shape(iris_dataset_labels), "\n")
```

```python
# train and test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(iris_dataset_features,iris_dataset_labels,test_size=0.20,random_state=42)
```

```python
#getting training data characteristics
print("x_train: \n", x_train, "\n")
print("np.shape(x_train): \n", np.shape(x_train), "\n")
print("y_train: \n", y_train, "\n")
print("np.shape(y_train): \n", np.shape(y_train), "\n")
```

```python
# Standard scaler preprocessing
from sklearn.preprocessing import StandardScaler 
#scale training and testing features
scaler=StandardScaler()
x_train_normalized=scaler.fit_transform(x_train)
x_test_normalized=scaler.transform(x_test)
```

```python
#Getting the normalized training data
print("x_train: \n",x_train_normalized,"\n")
print("np.shape(x_train): ",np.shape(x_train_normalized),"\n")
```

```python
#Getting the normalized test data
print("x_test: \n",x_test_normalized,"\n")
print("np.shape(x_test): ",np.shape(x_test_normalized),"\n")
```

```python
#Defining our SVM classifier
from sklearn.svm import SVC
svm_classifier=SVC(C=5.0,kernel="sigmoid")
print("SVM Classifier: \n", svm_classifier, "\n")
```

```python
from sklearn import svm
#Training our classifier
svm_classifier.fit(x_train_normalized,y_train)
#getting the classifier arguments
print("Printing the support vectors \n\n",svm_classifier.support_vectors_,"\n")
print("Number of support vectors \n\n",np.shape(svm_classifier.support_vectors_)[0],"\n")
#Yeah
print("Number of support vectors for each class: ",svm_classifier.n_support_)
print("Intercept values: ", svm_classifier.intercept_)
```

```python
#getting the predictions on train data
svm_train_predictions=svm_classifier.predict(x_train_normalized)
print("Predictions: \n",svm_train_predictions)
```

```python
#print accuracy
from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy Score is: ",accuracy_score(y_train,svm_train_predictions))
```

```python
#Comparing predictions with ground truth on training data
print("Predictions: \t Ground Truth: \n\n",np.column_stack((svm_train_predictions,y_train)))
```

```python
#Quantifiying the performance of the model on the test data
training_accuracy=svm_classifier.score(x_train_normalized,y_train)
print("Classifier Accuracy:: ", training_accuracy*100,"\n")
```

```python
#Getting the confusion matrix
from sklearn.metrics import confusion_matrix
train_cm=confusion_matrix(y_train,svm_train_predictions)
print("Confusion Matrix: \n\n",train_cm,"\n")
```

```python
import seaborn as sn 
sn.heatmap(train_cm,annot=True,cmap="OrRd")
```

```python
#Making predictions using the SVM model
svm_test_predictions=svm_classifier.predict(x_test_normalized)
print("Predictions: \n",svm_test_predictions)
```

```python
print("Predictions: \t Gound Truth: \n\n",np.column_stack((svm_test_predictions,y_test)),"\n")
```

```python
test_accuracy=svm_classifier.score(x_test_normalized,y_test)
print("classifier Accuracy:: ",test_accuracy*100,"\n")
```

```python
#Getting the confusion matrix predictions 
from sklearn.metrics import confusion_matrix
test_cm=confusion_matrix(y_test,svm_test_predictions)
print("Confusion Matrix: \n\n",test_cm,"\n")
import seaborn as sn 
sn.heatmap(test_cm,annot=True,cmap="OrRd")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,svm_test_predictions))
```
