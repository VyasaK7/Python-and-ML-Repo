```python
dataset1=[11,10,12,14,12,15,14,13,15,102,12,14,17,19,107,10,13,12,14,12,108,12,11,14,13,15,12,10,14,13,15,10]
outliers=[]
```

```python
def detect_outliers(dataset):
  threshold=3 #above 3rd standard deviation is considered outlier
  mean=np.mean(dataset)
  std=np.std(dataset)
  for i in dataset:
    z_score=(i-mean)/std
    if np.abs(z_score)>threshold:
      outliers.append(i)
  return outliers 
```

```python
outliers=detect_outliers(dataset1)
print(outliers)
#102 is not included bcoz 27*3=81 and 21 is myu(mean) so 104 or more will display as outlier
```
