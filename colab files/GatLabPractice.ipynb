```python
import sys
print("Checking Pallindrome")
no=input("Enter Number: ")
for i in range(0,len(no)):
  if int(no[i])==int(no[-1-i]):
    continue
  else:
    print("not pallindrome")
    sys.exit()
print("number is pallindrome")
```

```python
print("Swapping Variables")
a=10
b=20
print("before swap")
print(a)
print(b)
a=a+b
b=a-b
a=a-b
print("after swap")
print(a)
print(b)
```

```python
print("reverse of numbers")
st=input("enter string:")
st1=""
i=len(st)
i=i-2*i
for j in range(-1,i-1,-1):
    st1=st1+st[j]
    print(st[j],end="")
print("\n")
print(st1)
```
